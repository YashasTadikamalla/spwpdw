```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
x = np.random.rand(100,2)
kmeans = KMeans(n_clusters=3,random_state = 0)
```

```python
kmeans.fit(x)
```

```python
labels = kmeans.labels_
cluster_centers = kmeans.cluster_centers_
cluster_points = [[] for i in range(kmeans.n_clusters)]
for i , label in enumerate(labels):
  cluster_points[label].append(x[i])

```

```python
x1 = np.array(cluster_points[0])[:,0]
y1 = np.array(cluster_points[0])[:,1]

x2 = np.array(cluster_points[1])[:,0]
y2 = np.array(cluster_points[1])[:,1]

x3 = np.array(cluster_points[2])[:,0]
y3 = np.array(cluster_points[2])[:,1]

plt.scatter(x1,y1,label = "data0",edgecolor = "black")
plt.scatter(x2,y2,label= "data1",edgecolor = "black")
plt.scatter(x3,y3,label = "data2",edgecolor = "black")

# '''
all_points = np.vstack(cluster_points)
x_min, x_max = all_points[:, 0].min() - 1, all_points[:, 0].max() + 1
y_min, y_max = all_points[:, 1].min() - 1, all_points[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max,0.001), np.arange(y_min, y_max,0.001))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundary
plt.contourf(xx, yy, Z, alpha=0.5)
# '''

'''
from scipy.spatial import ConvexHull

# Draw convex hulls
def draw_convex_hull(points, color):
    hull = ConvexHull(points)
    for simplex in hull.simplices:
        plt.plot(points[simplex, 0], points[simplex, 1], color)

draw_convex_hull(np.array(cluster_points[0]), 'k-')
draw_convex_hull(np.array(cluster_points[1]), 'k-')
draw_convex_hull(np.array(cluster_points[2]), 'k-')
'''

plt.legend()
plt.xlabel("rand(x) --->")
plt.ylabel("rand(y) --->")
plt.title("KMeans clustering")
# plt.xlim(x_min, x_max)
# plt.ylim(y_min, y_max)
plt.show()
```
